```python
import numpy as np
# This dataset is dropped to the csv file followed by the tutorial in: https://archive.ics.uci.edu/dataset/144/statlog+german+credit+data

# Dataset Source:
# German credit data is a dataset by Hofmann, H. (1994). Statlog (German Credit Data) [Dataset]. UCI Machine Learning Repository. https://doi.org/10.24432/C5NC77.
from ucimlrepo import fetch_ucirepo
```

```python
# fetch dataset
statlog_german_credit_data = fetch_ucirepo(id=144)

dataset = statlog_german_credit_data.data
y = dataset.targets
```

```python
# data (as pandas dataframes)
X = dataset.features
X
```

```python
import pandas as pd
# 2. Extract features and targets as pandas dataframes
X = dataset.features
y = dataset.targets

# 3. Combine them into one single DataFrame
df = pd.concat([X, y], axis=1)
```

```python
df
```

```python
v = statlog_german_credit_data.variables
v
```

```python
mapping = dict(zip(v['name'], v['description']))
mapping['class'] = 'Credit granted'
mapping
```

```python
# 4. Rename the columns in the DataFrame
df_described = df.rename(columns=mapping)
```

```python
df_described
```

```python
df_described.dtypes
```

```python
encoding_map = {}
df_numeric = df_described.copy()
for column in df_numeric.columns:
    if df_numeric[column].dtype == 'str':
        df_numeric[column] = df_numeric[column].astype('category')

        # Store the mapping: {int_index: "string_label"}
        # Note: We store it as {int: str} for easy lookup later
        column_mapping = dict(enumerate(df_numeric[column].cat.categories))
        encoding_map[column] = {int(k): str(v) for k, v in column_mapping.items()}

        # Replace strings with codes
        df_numeric[column] = df_numeric[column].cat.codes
```

```python
encoding_map
```

```python
df_numeric
```

```python
df_numeric.loc[df_numeric['Credit granted'] == 2, 'Credit granted'] = 0     # 2 = bad
# df_numeric.loc[df_numeric['Credit granted'] == 1, 'Credit granted'] = 1     # 1 = good
```

```python
import json

# 3. Save the Numeric Dataset
df_numeric.to_csv("german_credit_numeric.csv", index=False)
```

```python
# 4. Save the Mapping (JSON)
with open("column_mapping.json", "w", encoding='utf-8') as f:
    json.dump(encoding_map, f, indent=4, ensure_ascii=False)
```
